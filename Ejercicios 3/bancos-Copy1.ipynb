```python
# Crea una nueva columna 'age_group'
df8=df7.copy()
df8['age_group'] = ''

# Recorre cada fila del dataset
for index, row in df7.iterrows():
    age = row['Age']
    
    # Verifica si la edad es mayor de 60 años
    if age > 60:
        df8.at[index, 'age_group'] = 'mayores de 60'
    else:
        df8.at[index, 'age_group'] = 'hasta 60'

# Muestra el dataset con la nueva columna
print(df8)
```
