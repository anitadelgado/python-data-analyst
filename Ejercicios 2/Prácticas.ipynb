```python
'''Ejercicio 3.10.1.
Dado el CSV, World_Development_Report_2021.csv, compartido por el tutor en este foro, crea un dataframe con los datos contenidos en él e indica cuál es el tipo de las variables que contiene'''

import pandas as pd

path = r'/Users/okashi/Desktop/World_Development_Report_2021.csv'

data = pd.read_csv(path,sep=',')

df1 = pd.DataFrame(data)
df1.info()
```

```python
'''Ejercicio 3.10.2.
Utilizando el dataframe que has creado en el ejercicio anterior, indica si hay filas duplicadas y, en ese caso, cuántas hay.'''

df1.duplicated().sum()
```

```python
'''Ejercicio 3.10.3.
En el caso de que hayas detectado filas duplicadas, elimínalas y guarda los resultados en el mismo dataframe en el que estás trabajando. 
Después de hacerlo, comprueba que ya no hay filas duplicadas.'''
#df.drop(columns=['Cabin'], inplace=True) # Machacamos el propio DataFrame

df1 = df1.drop_duplicates() # Machacamos el propio DataFrame
df1.duplicated().sum()
```

```python
'''Ejercicio 3.10.4.
En el dataframe con el que estamos trabajando, ¿hay alguna variable con elementos nulos? Y en el caso de que así sea, ¿qué variables son y cuántos nulos tiene cada una de ellas?'''

nulos = df1.isnull().sum()
nulos
```

```python
'''Ejercicio 3.10.5.
Utilizando el mismo dataframe que en los ejercicios anteriores, y teniendo en cuenta la variable percentage, ¿cuáles son el valor máximo, el valor mínimo y la media de dicha variable?'''

#df['percentage'].agg(['max', 'min', 'mean'])
print ('Valor máximo:', round(df1['percentage'].max(),2))
print ('Valor mínimo:', round(df1['percentage'].min(),2))
print ('Media:', round(df1['percentage'].mean(),2))
```

```python
'''Ejercicio 3.10.6.
¿Hay en el dataframe con el que estamos trabajando algún valor fuera de rango?'''

rango = df1.describe()
rango
```

```python
'''Ejercicio 3.10.8.
Obtén las frecuencias de las variables counted, population y percentage. ¿Las frecuencias son superiores al 2-5 %?'''

print ('Frecuencia counted:\n', round(df1['counted'].value_counts(normalize=True),2))
print ('Frecuencia population:\n', round(df1['population'].value_counts(normalize=True),2))
print ('Frecuencia percentage:\n', round(df1['percentage'].value_counts(normalize=True),2))
```
