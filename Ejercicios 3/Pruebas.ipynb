```python
import pandas as pd
from scipy.stats import chi2_contingency

# Crea una tabla de contingencia cruzando las variables 'Gender' y 'Exited'
contingency_table = pd.crosstab(df7['Gender'], df7['Exited'])

# Realiza la prueba de chi-cuadrado
chi2, p_value, _, _ = chi2_contingency(contingency_table)

# Imprime el resultado
print("Prueba de Chi-cuadrado:")
print("EstadÃ­stico de chi-cuadrado:", chi2)
print("Valor p:", p_value)
```

Esto significa que hay una asociaciÃ³n significativa entre las variables "Gender" (gÃ©nero) y "Exited" (abandono). El estadÃ­stico de chi-cuadrado indica la fuerza de la asociaciÃ³n, y un valor p muy pequeÃ±o (en este caso, prÃ¡cticamente cero) indica que la asociaciÃ³n es muy poco probable que se deba al azar.
En resumen, el anÃ¡lisis de la prueba de chi-cuadrado sugiere que el gÃ©nero estÃ¡ relacionado de manera significativa con el abandono de la entidad bancaria.

```python
contingency_table = pd.crosstab(df7['Geography'], df7['Exited'])

# Realiza la prueba de chi-cuadrado
chi2, p_value, _, _ = chi2_contingency(contingency_table)

# Imprime el resultado
print("Prueba de Chi-cuadrado:")
print("EstadÃ­stico de chi-cuadrado:", chi2)
print("Valor p:", p_value)
```

En este caso, el valor p es extremadamente pequeÃ±o (prÃ¡cticamente cero), lo que indica una asociaciÃ³n altamente significativa entre las variables "Geography"  y "Exited" . El estadÃ­stico de chi-cuadrado, en este caso, indica una fuerte evidencia en contra de la hipÃ³tesis nula de independencia entre estas variables.
En resumen, el anÃ¡lisis de la prueba de chi-cuadrado indica que el gÃ©nero estÃ¡ fuertemente asociado con el abandono de la entidad bancaria. 

```python
from scipy.stats import fisher_exact
contingency_table = pd.crosstab(df7['Gender'], df7['Exited'])


# Calcula el odds ratio y el intervalo de confianza utilizando la prueba exacta de Fisher
odds_ratio, p_value = fisher_exact(contingency_table)

# Imprime el resultado
print("Odds ratio:", odds_ratio)
print("Valor p:", p_value)
```

En este caso, el odds ratio es menor que 1, lo que indica que las personas del gÃ©nero femenino tienen un odds (riesgo) menor de abandonar la entidad bancaria en comparaciÃ³n con las personas del gÃ©nero masculino. AdemÃ¡s, el valor p es extremadamente pequeÃ±o, prÃ¡cticamente cero, lo que indica una asociaciÃ³n altamente significativa entre el gÃ©nero y el abandono de la entidad bancaria.

Estos resultados sugieren que el gÃ©nero tiene una influencia significativa en la propensiÃ³n a abandonar la entidad bancaria, y las personas del gÃ©nero masculino pueden tener un mayor riesgo de abandono en comparaciÃ³n con las personas del gÃ©nero femenino.

```python
import statsmodels.api as sm

# Ajustar el modelo de regresiÃ³n logÃ­stica
logit_model = sm.Logit(df7['Exited'], sm.add_constant(df7[['Gender', 'Geography']]))

# Obtener los resultados del modelo
result = logit_model.fit()

# Imprimir los resultados
print(result.summary())
```

Gender: El coeficiente asociado a la variable Gender es -0.5242. Esto significa que, manteniendo todas las demÃ¡s variables constantes, el gÃ©nero tiene un impacto significativo en la variable objetivo. En comparaciÃ³n con la categorÃ­a de referencia (que se elige automÃ¡ticamente por el modelo), los individuos masculinos tienen una probabilidad de abandono menor.

Geography: El coeficiente asociado a la variable Geography es 0.4255. Esto indica que, manteniendo todas las demÃ¡s variables constantes, la geografÃ­a tambiÃ©n tiene un impacto significativo en la variable objetivo.

AdemÃ¡s de los coeficientes estimados, la columna P>|z| representa los valores p asociados a cada coeficiente. Un valor p menor que 0.05 (o cualquier nivel de significancia previamente establecido) indica que el coeficiente es estadÃ­sticamente significativo

```python
#intento de nuevo poder usar las edades

# Crear las categorÃ­as de edad
bins = [18, 40, 60, 92]
labels = ['18-40','41-59', '60+']
df7['Age_Category'] = pd.cut(df7['Age'], bins=bins, labels=labels)
df7
```

```python
contingency_table = pd.crosstab(df7['Age_Category'], df7['Exited'])

# Realizar la prueba de chi-cuadrado
from scipy.stats import chi2_contingency
chi2, p_value, _, _ = chi2_contingency(contingency_table)

# Imprimir los resultados
print("EstadÃ­stico de chi-cuadrado:", chi2)
print("Valor p:", p_value)
```

Un valor de estadÃ­stico de chi-cuadrado alto y un valor p muy pequeÃ±o sugieren que la relaciÃ³n entre las categorÃ­as de edad y la variable "Exited" es significativa.

Esto significa que las categorÃ­as de edad tienen una influencia en la propensiÃ³n a abandonar la entidad bancaria. Puedes concluir que la edad es un factor relevante en la variable "Exited" y que las diferentes categorÃ­as de edad pueden tener tasas de abandono diferentes.

```python
# Calcular las frecuencias relativas de abandono por categorÃ­a de edad
abandoned_by_age = df7.groupby('Age')['Exited'].mean()

# Obtener el grupo de edad con la mayor proporciÃ³n de abandono
age_group_with_highest_abandonment = abandoned_by_age.idxmax()

# Imprimir los resultados
print("Grupo de edad con mayor proporciÃ³n de abandono:", age_group_with_highest_abandonment)
```
