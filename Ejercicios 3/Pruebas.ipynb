```python
import pandas as pd
from scipy.stats import chi2_contingency

# Crea una tabla de contingencia cruzando las variables 'Gender' y 'Exited'
contingency_table = pd.crosstab(df7['Gender'], df7['Exited'])

# Realiza la prueba de chi-cuadrado
chi2, p_value, _, _ = chi2_contingency(contingency_table)

# Imprime el resultado
print("Prueba de Chi-cuadrado:")
print("Estadístico de chi-cuadrado:", chi2)
print("Valor p:", p_value)
```

Esto significa que hay una asociación significativa entre las variables "Gender" (género) y "Exited" (abandono). El estadístico de chi-cuadrado indica la fuerza de la asociación, y un valor p muy pequeño (en este caso, prácticamente cero) indica que la asociación es muy poco probable que se deba al azar.
En resumen, el análisis de la prueba de chi-cuadrado sugiere que el género está relacionado de manera significativa con el abandono de la entidad bancaria.

```python
contingency_table = pd.crosstab(df7['Geography'], df7['Exited'])

# Realiza la prueba de chi-cuadrado
chi2, p_value, _, _ = chi2_contingency(contingency_table)

# Imprime el resultado
print("Prueba de Chi-cuadrado:")
print("Estadístico de chi-cuadrado:", chi2)
print("Valor p:", p_value)
```

En este caso, el valor p es extremadamente pequeño (prácticamente cero), lo que indica una asociación altamente significativa entre las variables "Geography"  y "Exited" . El estadístico de chi-cuadrado, en este caso, indica una fuerte evidencia en contra de la hipótesis nula de independencia entre estas variables.
En resumen, el análisis de la prueba de chi-cuadrado indica que el género está fuertemente asociado con el abandono de la entidad bancaria. 

```python
from scipy.stats import fisher_exact
contingency_table = pd.crosstab(df7['Gender'], df7['Exited'])


# Calcula el odds ratio y el intervalo de confianza utilizando la prueba exacta de Fisher
odds_ratio, p_value = fisher_exact(contingency_table)

# Imprime el resultado
print("Odds ratio:", odds_ratio)
print("Valor p:", p_value)
```

En este caso, el odds ratio es menor que 1, lo que indica que las personas del género femenino tienen un odds (riesgo) menor de abandonar la entidad bancaria en comparación con las personas del género masculino. Además, el valor p es extremadamente pequeño, prácticamente cero, lo que indica una asociación altamente significativa entre el género y el abandono de la entidad bancaria.

Estos resultados sugieren que el género tiene una influencia significativa en la propensión a abandonar la entidad bancaria, y las personas del género masculino pueden tener un mayor riesgo de abandono en comparación con las personas del género femenino.

```python
import statsmodels.api as sm

# Ajustar el modelo de regresión logística
logit_model = sm.Logit(df7['Exited'], sm.add_constant(df7[['Gender', 'Geography']]))

# Obtener los resultados del modelo
result = logit_model.fit()

# Imprimir los resultados
print(result.summary())
```

Const: Este coeficiente representa el término de intercepción de la regresión logística. En este caso, el valor estimado es -1.6805.

Gender: El coeficiente asociado a la variable Gender es -0.5242. Esto significa que, manteniendo todas las demás variables constantes, el género tiene un impacto significativo en la variable objetivo. En comparación con la categoría de referencia (que se elige automáticamente por el modelo), los individuos masculinos tienen una probabilidad de abandono menor.

Geography: El coeficiente asociado a la variable Geography es 0.4255. Esto indica que, manteniendo todas las demás variables constantes, la geografía también tiene un impacto significativo en la variable objetivo.

Además de los coeficientes estimados, la columna P>|z| representa los valores p asociados a cada coeficiente. Un valor p menor que 0.05 (o cualquier nivel de significancia previamente establecido) indica que el coeficiente es estadísticamente significativo

```python
#intento de nuevo poder usar las edades

# Crear las categorías de edad
bins = [18, 40, 60, 92]
labels = ['18-40','41-59', '60+']
df7['Age_Category'] = pd.cut(df7['Age'], bins=bins, labels=labels)
df7
```

```python
contingency_table = pd.crosstab(df7['Age_Category'], df7['Exited'])

# Realizar la prueba de chi-cuadrado
from scipy.stats import chi2_contingency
chi2, p_value, _, _ = chi2_contingency(contingency_table)

# Imprimir los resultados
print("Estadístico de chi-cuadrado:", chi2)
print("Valor p:", p_value)
```

Un valor de estadístico de chi-cuadrado alto y un valor p muy pequeño sugieren que la relación entre las categorías de edad y la variable "Exited" es significativa.

Esto significa que las categorías de edad tienen una influencia en la propensión a abandonar la entidad bancaria. Puedes concluir que la edad es un factor relevante en la variable "Exited" y que las diferentes categorías de edad pueden tener tasas de abandono diferentes.

```python
# Calcular las frecuencias relativas de abandono por categoría de edad
abandoned_by_age = df7.groupby('Age')['Exited'].mean()

# Obtener el grupo de edad con la mayor proporción de abandono
age_group_with_highest_abandonment = abandoned_by_age.idxmax()

# Imprimir los resultados
print("Grupo de edad con mayor proporción de abandono:", age_group_with_highest_abandonment)
```
