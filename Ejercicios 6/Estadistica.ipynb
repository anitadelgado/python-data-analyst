```python
'''Ejercicio 3.4.1.
puntuaciones de los exámenes 5, 3, 6, 5, 4, 5, 7, 8, 6, 5, 6, 8, 3, 4, 5, 4, 8, 5, 5, 4. Calcula los siguientes estadísticos:

a)    La media
b)    La mediana
c)    Los cuartiles Q1, Q2 y Q3
d)    Los percentiles P30 y P80
e)    La desviación típica
f)    La varianza
g)    El rango intercuartílico'''

import numpy as np
notas = np.array([5, 3, 6, 5, 4, 5, 7, 8, 6, 5, 6, 8, 3, 4, 5, 4, 8, 5, 5, 4])

print('La media de las notas es', np.mean(notas))
print('La mediana de las notas es', np.median(notas))
print('Los cuartiles Q1,Q2 Y Q3 son:', np.quantile(notas,0.25),',',np.quantile(notas,0.50),'y',np.quantile(notas,0.75))
print('Los percentiles P30 y P80 son:',np.percentile(notas,30),'y',np.percentile(notas,80))
print('La desviación típica es:',np.std(notas))
print('La varianza es:', np.var(notas))
print('El rango intercuartílico es:',np.quantile(notas,0.75)-np.quantile(notas,0.25))
```

```python
'''Ejercicio 3.4.2.
Dado el siguiente diccionario de datos ['D_ejer1':[5, 3, 6, 5], 'D_ejer2’:[4, 5, 7, 8]]
a)    La media
b)    La mediana
c)    Los cuartiles Q2 y Q3
d)    Los percentiles P40 y P90
e)    La desviación típica
f)    La varianza
g)    La covarianza
h)    La correlación de Pearson
i)    La matriz de correlación de Pearson'''

import pandas as pd
import seaborn

datos = pd.DataFrame({'D_ejer1':[5,3,6,5],'D_ejer2':[4,5,7,8]})
corr_matrix = datos.corr(method = 'pearson') 

print('MEDIA:\n',datos.mean(),'\n')
print('MEDIANA:\n',datos.median(),'\n')
print('CUARTILES Q2 y Q3:\n',datos.quantile([0.5,0.75]),'\n')
print('PERCENTILES P40 y P90:\n',datos.quantile([0.4,0.9]),'\n')
print('DESVIACIÓN TÍPICA:\n',datos.std(),'\n')
print('VARIANZA:\n',datos.var(),'\n')
print('COVARIANZA:\n',datos.cov(),'\n')
print('CORRELACIÓN DE PEARSON:\n',datos.corr(),'\n')
print('MATRIZ:\n',seaborn.heatmap(corr_matrix, annot= True, cmap = 'coolwarm'))
```

```python
'''Ejercicio 3.4.7.
Dada la siguiente tabla:
Crea un dataframe con los datos que hay en ella y responde a las siguientes preguntas:

a)     ¿Cuál es la media de facturación de los artículos de la tabla?

b)    ¿Cuál es la varianza de los pedidos?

c)     ¿Cuál es la covarianza entre los pedidos y la facturación?

d)    ¿Cuál es la media de la facturación y los pedidos entregados?'''

import pandas as pd

tabla = {'Articulos':['Art0001','Art0002','Art0003','Art0004','Art0005'],
         'Pedidos':[325,125,205,54,186],
         'Entregado':['S','S','N','N','S'],
         'Facturación':[8125,9375,13325,18900,9300]}

df = pd.DataFrame(tabla)
pedidos_entregados = df[df['Entregado'] == 'S']

print('Media de facturación:',(df['Facturación'].mean()))
print('Varianza de pedidos:', (df['Pedidos'].var()))
print('La covarianza entre pedidos:', df['Pedidos'].cov(df['Facturación']))
print('La media de facturación de los pedidos entregados:',pedidos_entregados['Facturación'].mean())
```

```python
'''Ejercicio 3.5.1.
Si tienes la siguiente información sobre las notas de un grupo de alumnos tanto en las asignaturas de lengua como de matemáticas:
Notas en lengua: 5,7,6,2,8,9,3,4,1,9,5,4,6,4,8,7,6,2,8,9
Notas en matemáticas: 7,8,5,6,8,3,2,9,9,6,1,5,7,3,5,4,5,9,7,6
Representa esta información en un gráfico de dispersión. ¿Existe alguna correlación entre las variables?'''

import matplotlib.pyplot as plt 
import numpy as np

lengua=(5,7,6,2,8,9,3,4,1,9,5,4,6,4,8,7,6,2,8,9)
matemáticas=(7,8,5,6,8,3,2,9,9,6,1,5,7,3,5,4,5,9,7,6)
plt.scatter(lengua,matemáticas)
plt.xlabel('lengua')
plt.ylabel('matemáticas')
#plt.plot([0,9]),([0,9])
plt.show()

correlation = np.corrcoef(lengua, matemáticas)[0, 1]
print('Correlación:', correlation)
```
