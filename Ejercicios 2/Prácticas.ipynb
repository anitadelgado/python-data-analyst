```python
'''Ejercicio 3.10.1.
Dado el CSV, World_Development_Report_2021.csv, compartido por el tutor en este foro, crea un dataframe con los datos contenidos en √©l e indica cu√°l es el tipo de las variables que contiene'''

import pandas as pd

path = r'/Users/okashi/Desktop/World_Development_Report_2021.csv'

data = pd.read_csv(path,sep=',')

df1 = pd.DataFrame(data)
df1.info()
```

```python
'''Ejercicio 3.10.2.
Utilizando el dataframe que has creado en el ejercicio anterior, indica si hay filas duplicadas y, en ese caso, cu√°ntas hay.'''

df1.duplicated().sum()
```

```python
'''Ejercicio 3.10.3.
En el caso de que hayas detectado filas duplicadas, elim√≠nalas y guarda los resultados en el mismo dataframe en el que est√°s trabajando. 
Despu√©s de hacerlo, comprueba que ya no hay filas duplicadas.'''
#df.drop(columns=['Cabin'], inplace=True) # Machacamos el propio DataFrame

df1 = df1.drop_duplicates() # Machacamos el propio DataFrame
df1.duplicated().sum()
```

```python
'''Ejercicio 3.10.4.
En el dataframe con el que estamos trabajando, ¬øhay alguna variable con elementos nulos? Y en el caso de que as√≠ sea, ¬øqu√© variables son y cu√°ntos nulos tiene cada una de ellas?'''

nulos = df1.isnull().sum()
nulos
```

```python
'''Ejercicio 3.10.5.
Utilizando el mismo dataframe que en los ejercicios anteriores, y teniendo en cuenta la variable percentage, ¬øcu√°les son el valor m√°ximo, el valor m√≠nimo y la media de dicha variable?'''

#df['percentage'].agg(['max', 'min', 'mean'])
print ('Valor m√°ximo:', round(df1['percentage'].max(),2))
print ('Valor m√≠nimo:', round(df1['percentage'].min(),2))
print ('Media:', round(df1['percentage'].mean(),2))
```

```python
'''Ejercicio 3.10.6.
¬øHay en el dataframe con el que estamos trabajando alg√∫n valor fuera de rango?'''

rango = df1.describe()
rango
```

```python
'''Ejercicio 3.10.8.
Obt√©n las frecuencias de las variables counted, population y percentage. ¬øLas frecuencias son superiores al 2-5 %?'''

print ('Frecuencia counted:\n', round(df1['counted'].value_counts(normalize=True),2))
print ('Frecuencia population:\n', round(df1['population'].value_counts(normalize=True),2))
print ('Frecuencia percentage:\n', round(df1['percentage'].value_counts(normalize=True),2))
```
