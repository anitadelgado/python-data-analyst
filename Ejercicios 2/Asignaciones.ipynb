```python
'''Ejercicio 3.9.1.
Crea tabla  ordénalos de forma ascendente por la columna Facturacion'''


import pandas as pd

datos = {'Mes':['Enero','Febrero','Marzo','Abril','Mayo','Junio','Julio','Agosto','Septiembre','Octubre','Noviembre','Diciembre'],
         'Pedidos': [20,40,None,25,10,35,None,None,22,42,36,None],
         'Facturacion': [15000,30000,2500,None,25000,30000,1500,None,15000,35000,25000,2500],
         'Gastos': [5000,6200,4500,3100,5000,5500,6000,5000,6500,5250,7000,5200],
         'Pendientes': [None,None,25,None,6,5,10,None,None,None,4,None],
         'En_revision': [None,None,None,None,3,15,2,None,1,None,None,1]}

df_fac = pd.DataFrame(datos)

df_fac = df_fac.sort_values(by='Facturacion')
df_fac
         
```

```python
'''Ejercicio 3.9.2.
Ordena el dataframe anterior por gastos descendentes'''

df_fac = df_fac.sort_values(by=['Gastos'],ascending = False)
df_fac
```

```python
'''Ejercicio 3.9.3.
Ordena el dataframe anterior por Gastos y por Facturación. Y también por Facturación y por Gastos. ¿El resultado es el mismo?'''

df_fac = df_fac.sort_values(by=['Facturacion','Gastos'])
df_fac
```

```python
'''Ejercicio 3.9.4.
Ordena el dataframe en función de su índice y modifica el nombre de la columna Facturacion por Facturas'''

df_fac1 = df_fac.copy()
df_fac1 = df_fac1.rename(columns = {'Facturacion':'Facturas'})
df_fac1.sort_index()
```

```python
'''Ejercicio 3.9.5.
¿Cuántos elementos nulos tiene cada una de las columnas del dataframe con el que estamos trabajando?'''

df_fac1.isnull().sum()
```

```python
'''Ejercicio 3.9.6.
¿Cuál es la media de valores nulos en cada una de las columnas del dataframe?'''

df_fac1.isnull().mean().round(2)
```

```python
'''Ejercicio 3.9.7.
¿Qué meses no tienen pedidos pendientes? Y ¿en qué meses no ha habido pedidos?'''

df_fac2 = df_fac1.sort_index()

print('Sin pedidos pendientes:\n',df_fac2[df_fac2['Pendientes'].isnull()]['Mes'],'\n')
print('Sin pedidos:\n',df_fac2[df_fac2['Pedidos'].isnull()]['Mes'])
```

```python
'''Ejercicio 3.9.8.
¿Hay algún mes en el que no hay ningún valor nulo?'''

df_fac3 = df_fac2.dropna()
df_fac3
```
