```python
'''Ejercicio 3.15.1.
Utilizando el dataset breast_cancer de sklearn, 
aplica la técnica K-Fold, con 3 splits, con la métrica accuracy y muestra el cross val score obtenido con el modelo de regresión logística'''

from sklearn.model_selection import KFold, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn import datasets
import pandas as pd
import numpy as np

breast_cancer = datasets.load_breast_cancer()
x = breast_cancer.data[:,:]
y = breast_cancer.target

kfold = KFold(n_splits=3)
model = LogisticRegression(max_iter=2500)
scoring='accuracy'

results=cross_val_score(model,x,y,cv=kfold,scoring = scoring)

print('Cross val score:',results)
print('La media es:', round(results.mean(),2))

```

```python
'''Ejercicio 3.16.5.
Utilizando el dataset iris de sklearn, 
parametriza el algoritmo para un modelo de árbol de decesión con la técnica Grid Search utilizando los siguientes datos

model = DecisionTreeClassifier()

param_grid = {"max_depth":[4], "criterion":["entropy"]}

¿Cuáles son la puntuación del modelo y la mejor configuración? ¿Y cómo interpretas esa puntuación?'''

import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier

iris = datasets.load_iris()
X = iris.data
y = iris.target

model = DecisionTreeClassifier()
param_grid = {'max_depth':[4], 'criterion':['entropy']}
rsearch = GridSearchCV (estimator=model, param_grid=param_grid)
rsearch.fit(X,y)

print('Modelo:', rsearch.score(X,y).round(2))
print('Mejor puntuación:', rsearch.best_score_.round(2))
print('Configuración:', rsearch.best_params_)
```

```python
'''Ejercicio 3.16.6.
Y si lo parametrizas utilizando Randomized Search CV, ¿cuáles son la puntuación del modelo 
la mejor configuración con las mismas características dadas en el ejercicio anterior?'''

import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn.model_selection import RandomizedSearchCV
from sklearn.tree import DecisionTreeClassifier


iris = datasets.load_iris()
X = iris.data
y = iris.target

model = DecisionTreeClassifier()
param_grid = {'max_depth':[4], 'criterion':['entropy']}
rsearch = RandomizedSearchCV (model, param_grid, n_iter=1)
rsearch.fit(X,y)

print('Modelo:', rsearch.score(X,y).round(2))
print('Mejor puntuación:', rsearch.best_score_.round(2))
print('Configuración:', rsearch.best_params_)
```

```python
'''Ejercicio 3.17.1.
Dado el dataset credictcard2, compartido por el tutor en este foro, balancea los datos ajustando los parámetros del modelo.'''
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

path = r'/Users/okashi/Desktop/df_creditcard2.csv'
df = pd.read_csv(path, sep=',')
df

LABELS = ['Normal','Fraud']
X=df.drop('Class',axis=1)
y=df['Class']

X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.8)

#Definimos las funciones que vamos a necesitar #Para balancear los datos
def run_model_balanced (X_train, X_test, y_train, y_test):
    clf = LogisticRegression (C = 1.0, penalty = 'l2', random_state = 1, solver = "newton-cg", class_weight = "balanced")
    clf.fit(X_train,y_train)
    return clf

#Para mostrar los resultados
def mostrar_resultados(y_test, pred_y):
    conf_matrix = confusion_matrix(y_test, pred_y)
    plt.figure(figsize=(8, 8))
    sns.heatmap(conf_matrix, xticklabels=LABELS, yticklabels=LABELS, annot=True, fmt="d") 
    plt.title("Confusion matrix")
    plt.ylabel('True class')
    plt.xlabel('Predicted class')
    plt.show()
    print (classification_report(y_test, pred_y))
#Entrenamos y testeamos
model = run_model_balanced (X_train,X_test,y_train,y_test) 
pred_y = model.predict(X_test)
mostrar_resultados(y_test, pred_y)
```

```python
'''Ejercicio 3.17.2.
Utilizando los mismos datos, ¿qué resultado obtienes utilizando el algoritmo NearMiss?'''

#Importamos las librerías específicas que no habíamos importado anteriormente
from collections import Counter
from imblearn.under_sampling import NearMiss

#Definimos las variables que no estaban definidas
us = NearMiss(n_neighbors=3,version=2)
X_train_res, y_train_res = us.fit_resample(X_train, y_train)
#Definimos la función que vamos a necesitar para balancear los datos sin el class_weight
def run_model (X_train, X_test, y_train, y_test):
    clf = LogisticRegression(C=1.0,penalty='l2',random_state=1,solver="newton-cg") 
    clf.fit(X_train,y_train)
    return clf
#Mostramos los valores de antes de balancear los datos y de después
print("Distribución antes del balanceo {}".format(Counter(y_train))) 
print("Distribución después del balanceo {}".format(Counter(y_train_res))) #Entrenamos y testeamos

model = run_model(X_train_res, X_test, y_train_res, y_test)
pred_y = model.predict(X_test)
mostrar_resultados(y_test, pred_y)
```
