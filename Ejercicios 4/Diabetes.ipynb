```python
import matplotlib.pyplot as plt

female_data = df[df['gender'] == 'female']
male_data = df[df['gender'] == 'male']

# Crear los gráficos de densidad
plt.figure(figsize=(8, 6))
female_data['diabetes'].plot(kind='density', label='Female')
male_data['diabetes'].plot(kind='density', label='Male')

# Personalizar el gráfico
plt.xlabel('Diabetes')
plt.ylabel('Density')
plt.title('Density Plot of Diabetes by Gender')
plt.legend()

# Mostrar el gráfico
plt.show()

```

```python
import numpy as np

 

columnas_outliers = ["bmi","Blood_glucose_level", "HBA1c_level"]  

 

for columna in columnas_outliers:
    Q1 = np.quantile(df[columna], 0.25)
    Q3 = np.quantile(df[columna], 0.75)
    IQR = Q3 - Q1

 

    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR

 

    df[columna] = np.clip(df[columna], lower_limit, upper_limit)
```
