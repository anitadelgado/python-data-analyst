```python
'''Ejercicio 3.8.1
Crea un dataframe con los siguientes datos
Inserta una columna que se llame Dividendos y que sea igual a la columna Beneficio.'''

import pandas as pd

datos = {'Mes':['Enero','Febrero','Marzo','Abril','Mayo','Junio','Julio','Agosto','Septiembre','Octubre','Noviembre','Diciembre'],
         'Ventas':[10000, 15200, 14300, 25800, 2560, 12780, 32500, 23624, 12000, 15858, 23250, 11630],
         'Gastos':[6500, 6500, 6500, 7850, 6500, 8250, 6500, 9252, 6500, 10250, 12450, 13000],
         'Beneficio':[3500,8700,7800,17950,-3940,4530,26000,14372,5500,5608,10800,-1370],
         'Cobrado':['S', 'N', 'N', 'S', 'S', 'N', 'S', 'S', 'S', 'N', 'S', 'N']}

df = pd.DataFrame(datos)
df = df.assign(Dividendos = df['Beneficio'])
df
```

```python
'''Ejercicio 3.8.2.
Utilizando el dataframe anterior, modifica la columna Dividendo para que contenga el 10% de su valor actual utilizando el método assign'''

df = df.assign(Dividendos = df['Dividendos']*0.1)
df
```

```python
'''Ejercicio 3.8.3.
En el resultado obtenido en el ejercicio anterior, verás que hay meses en los que los dividendos son negativos, 
como esto no puede suceder, modifica la columna Dividendo para que cuando el resultado sea positivo se muestre pero que cuando sea negativo ponga el valor 0. Utiliza la función map.'''

df['Dividendos']=df['Dividendos'].map(lambda x: x if x > 0 else 0)
df
```

```python
'''Ejercicio 3.8.4.
Multiplica por dos, utilizando la función map, el valor de todos los elementos de la columna Gastos del dataframe.'''

df1 = df.copy()
df1['Gastos']=df1['Gastos'].map(lambda x: x*2)
df1
```

```python
'''Ejercicio 3.8.5.
Recalcula el beneficio y los dividendos mensuales teniendo en cuenta los gastos modificados en el ejercicio anterior.'''
df2=df1.copy()
df2['Beneficio']=df2['Ventas']-df2['Gastos']
df2['Dividendos']=df2['Beneficio']*0.1
df2['Dividendos']=df2['Dividendos'].map(lambda x: x if x > 0 else 0)
df2
```

```python
'''Ejercicio 3.8.6.
Incluye una columna en el dataframe que muestra cuál es el porcentaje que supone el Dividendo obtenido sobre las Ventas realizadas. Utiliza la función apply para incluir la columna.
NOTA: La fórmula para calcular el porcentaje será (Dividendo/Ventas)*100'''

def porcentaje (num):
    return (num['Dividendos']/num['Ventas'])*100

df2["Porcentaje"] = round(df2.apply(porcentaje, axis=1))
df2
```

```python
'''Ejercicio 3.8.7.
Crea un nuevo dataframe que contenga solo las columnas que contienen texto. Utiliza la función drop para conseguirlo.'''

df3 = df2.copy()
df3_tipos = df3.dtypes
print(df3_tipos)
```

```python
df3 = df3.drop(columns=['Ventas','Gastos','Beneficio','Dividendos','Porcentaje'])
df3
```

```python
'''Ejercicio 3.8.8.
Muestra la media de todos los datos de las columnas numéricas agrupados por si están cobrados o no.'''

media = df2.groupby('Cobrado').mean()
media
```
